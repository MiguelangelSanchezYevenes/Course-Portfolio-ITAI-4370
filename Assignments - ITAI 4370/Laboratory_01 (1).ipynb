```python
import numpy as np
import matplotlib.pyplot as plt

bits = np.array([1, 0, 1, 1, 0, 0, 1, 0])
bpsk = 2 * bits - 1

np.random.seed(1)
received = bpsk + 0.4 * np.random.randn(len(bpsk))
decoded = (received >= 0).astype(int)

print("Source Bits:     ", bits)
print("BPSK Signal:     ", bpsk)
print("Received Signal: ", np.round(received, 2))
print("Receiver Bits:   ", decoded)

plots = [
    (bits, "Source Bits", "Bit Value", "step"),
    (bpsk, "BPSK Modulated Signal", "Amplitude", "stem"),
    (received, "Received Signal After Channel Noise", "Amplitude", "stem"),
    (decoded, "Receiver Bits After Demodulation", "Bit Value", "step")
]

for data, title, ylabel, kind in plots:
    plt.figure()
    if kind == "step":
        plt.step(range(len(data)), data, where="mid")
    else:
        plt.stem(range(len(data)), data)
    if "Received" in title:
        plt.axhline(0, linestyle="--")
    plt.title(title)
    plt.xlabel("Bit/Symbol Index")
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()
```
